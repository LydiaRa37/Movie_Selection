```python
# Importation des modules
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import re

# Modifier les options d'affichage pour limiter les décimales
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Affichage de la BDD
TmdbCleanDumActor_Final = pd.read_csv('BDD_CSV/last_BDD/TmdbCleanDumActor_Final.csv')
print(TmdbCleanDumActor_Final.info())
```

```python
# Affichage
print(TmdbCleanDumActor_Final.sample(1).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Initialisation des comptes de NaN
nan_counts_per_column = {}
total_nan_counts = 0

# Boucle sur chaque colonne et comptage des NaN
for column in TmdbCleanDumActor_Final.columns:
    nan_count = TmdbCleanDumActor_Final[column].isna().sum()
    nan_counts_per_column[column] = nan_count
    total_nan_counts += nan_count

# Affichage des résultats
print("Nombre de NaN par colonne:")
for column, count in nan_counts_per_column.items():
    print(f"Colonne {column}: {count} NaN")
```

```python
# Affichage de la BDD
df_netflix = pd.read_csv('BDD_CSV/last_BDD/netflix.csv')
print(df_netflix.info())
print(df_netflix.sample(2).to_markdown(index=False, numalign="left", stralign="left"))
```

#### ***Merge de TmdbCleanDumActor_Final et netflix***

```python
df_final = TmdbCleanDumActor_Final.merge(right= df_netflix, on="ID_title", how='left')
```

```python
print(df_final.info())
print(df_final.sample(2).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Remplacer les NaN dans 'original_language_x' par les valeurs correspondantes de 'original_language_y' provenant de netflix
df_final['original_language_x'] = df_final['original_language_x'].fillna(df_final['original_language_y'])
```

```python
# Suppression des colonnes avec le suffixe `_y`
df_final = df_final.loc[:, ~df_final.columns.str.endswith('_y')]

# Renommage des colonnes en enlevant le suffixe `_x`
df_final.columns = df_final.columns.str.replace('_x$', '', regex=True)
```

```python
df_final.columns
```

```python
# Initialisation d'un dictionnaire pour stocker le nombre de doublons par colonne
duplicate_counts = {}

# Pour chaque colonne du DataFrame
for col in df_final.columns:
    # Utilisation de duplicated() pour détecter les doublons
    num_duplicates = df_final[col].duplicated(keep=False).sum() - df_final[col].duplicated(keep='first').sum()
    duplicate_counts[col] = num_duplicates

# Affichage du nombre de doublons par colonne
for col, count in duplicate_counts.items():
    print(f"Colonne '{col}' a {count} doublons")
```

```python
# Initialisation des comptes de NaN
nan_counts_per_column = {}
total_nan_counts = 0

# Boucle sur chaque colonne et comptage des NaN
for column in df_final.columns:
    nan_count = df_final[column].isna().sum()
    nan_counts_per_column[column] = nan_count
    total_nan_counts += nan_count

# Affichage des résultats
print("Nombre de NaN par colonne:")
for column, count in nan_counts_per_column.items():
    print(f"Colonne {column}: {count} NaN")
```

```python
# Supprimer les lignes doublons
df_final = df_final.drop_duplicates()

#  Réinitialiser l'index après suppression des doublons
df_final = df_final.reset_index(drop=True)
```

```python
print(df_final.info())
print(df_final.sample(2).to_markdown(index=False, numalign="left", stralign="left"))
```

#### ***Selection des TOP acteurs/actrices***

```python
# Filtrer les colonnes nécessaires
df_filtered = df_final[['ID_name', 'actor_name', 'weighted_averageRating']]

# Calculer la moyenne pondérée et le nombre de films pour chaque acteur
average_ratings = df_filtered.groupby('ID_name').agg(
    average_weighted_rating=('weighted_averageRating', 'mean'),  # Calculer la moyenne pondérée
    actor_name=('actor_name', 'first'),                          # Prendre le premier acteur associé à l'ID_name
    number_of_films=('actor_name', 'size')                       # Compter le nombre de films
).reset_index()

# Réorganiser les colonnes : 
average_ratings = average_ratings[['ID_name', 'actor_name', 'average_weighted_rating', 'number_of_films']]
```

```python
average_ratings.info()
```

```python
# Trier le DataFrame par 'number_of_films' en ordre décroissant
average_ratings_sorted = average_ratings.sort_values(by='number_of_films', ascending=False)

# Prendre un échantillon de ... lignes du DataFrame trié et l'afficher en format markdown
print(average_ratings_sorted.head(50).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
# Filtrer pour obtenir uniquement les ID_name où 'number_of_films' est sup à une certaine valeur 
filtered_id_names = average_ratings[average_ratings['number_of_films'] > 3]['ID_name']
```

```python
# Convertir le résultat en liste (si nécessaire)
id_name_list = filtered_id_names.tolist()
len(id_name_list)
```

```python
# Récupérer les lignes du DataFrame TmdbCleanDumActor_Final où ID_name est dans id_name_list
filtered_df2 = df_final[TmdbCleanDumActor_Final['ID_name'].isin(id_name_list)]
filtered_df2.info()
```

### Dummies

```python
# Creating dummy variables from the 'actor_name' column
df_dum2 = pd.get_dummies(filtered_df2['actor_name'])

# Concaténer les variables factices au DataFrame d'origine (facultatif)
df = pd.concat([filtered_df2, df_dum2], axis=1)
```

```python
df.info()
```

```python
# Boucle pour imprimer chaque colonne et son type de données
for column, dtype in df.dtypes.items():
    print(f'{column}: {dtype}')
```

```python
# Identifier les colonnes de type bool
bool_columns = df.select_dtypes(include='bool').columns

# Convertir les colonnes bool en int
df[bool_columns] = df[bool_columns].astype(int)
```
